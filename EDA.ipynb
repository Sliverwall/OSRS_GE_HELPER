# Explore datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
```

# Margin EDA

```python
# Read margins csv as df. Adjust unix timestamps to make it easier to read
margins = pd.read_csv("data/runelite/runelite_data_margins.csv")
margins['highTime'] = pd.to_datetime(margins['highTime'], unit='s')
margins['lowTime'] = pd.to_datetime(margins['lowTime'], unit='s')
margins.head()
```

```python
# Read in item keys to match limits to given items
itemKeys = pd.read_csv("data/item_keys.csv")
itemKeys.head()
```

```python
# Merge the DataFrames on ids
# First, filter df2 to keep only the join key and the desired column
filteredItemKeys = itemKeys[['id', 'limit']]

# Now merge
merged_df = pd.merge(
    margins, filteredItemKeys,
    left_on='itemID',
    right_on='id',
    how='left'
)
# Drop the extra id column
merged_df = merged_df.drop(columns='id')
merged_df.head()
```

```python
'''Find high volumne, high margin items'''
# Filter out rows where limit is less than 10,000
highVolumeDF = merged_df[merged_df['limit'] >= 10000]

highVolumeDF = highVolumeDF[highVolumeDF['high'] <= 1000000]

# Sort by marginAdjusted in descending order
highVolumeDF = highVolumeDF.sort_values(by='marginAdjusted', ascending=False)

highVolumeDF.head()
```

```python
'''Find low volumne, high margin items'''
lowVolumeDF = merged_df[merged_df['high'] <= 100000000]
# Sort by marginAdjusted in descending order
lowVolumeDF = lowVolumeDF.sort_values(by='marginAdjusted', ascending=False)

lowVolumeDF.head()
```

# Time Series EDA for specific items

```python
itemID = 3486
df = pd.read_csv(f"data/runelite/runelite_data_5m_{itemID}.csv")

# Get name of item
itemName = df['name'][0]
# Convert the 'timestamp' column to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
df.head()
```

```python
# Create the plot with an appropriate figure size
plt.figure(figsize=(10, 5))

# Plot the data series
plt.plot(df['timestamp'], df['avgHighPrice'], label='Average High Price', marker='o')
plt.plot(df['timestamp'], df['avgLowPrice'], label='Average Low Price', marker='o')

# Add title and labels
plt.title(f'{itemName} Price Data Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Price')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Format the x-axis to display day and time down to the minute
date_formatter = mdates.DateFormatter('%Y-%m-%d %H:%M')
plt.gca().xaxis.set_major_formatter(date_formatter)

# Format the y-axis with comma separators for better readability.
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:,.0f}'))

# Optionally, you might adjust the tick frequency on the x-axis (example: every 5 minutes)
# plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=5))

# Add legend and grid for clarity
plt.legend()
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```
