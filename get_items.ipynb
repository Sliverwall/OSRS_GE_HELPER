# get_items

- Uses runelite's API to get a mapping for each item id and item name + other basic features

```python
import requests
import json
import csv
import time
import os.path
```

```python
'''Define needed variables'''
# API key used to access summary jsons
runeliteAPI_mapping = "https://prices.runescape.wiki/api/v1/osrs/mapping"

# User agent tag used to let wiki know why I am using their API. 
headers = {
    'User-Agent': 'Machine Learning Practice',
    'From': 'sliverwall (discord ID)'  # discord ID
}
```

```python
'''Get data from api'''
# Define headers for output file
headers = ["id", "name", "limit", "highalch"]
labels = []
# Get API request and load it into json format
r = requests.get(runeliteAPI_mapping)
json_data = json.loads(r.text)

# Loop through every item in the mapping then extract needed labels
for item in json_data:
    # Only take items with defined limits
    if "limit" in item and "highalch" in item:
        entry = (
            item["id"],
            item["name"].replace(" ", "_"),
            item["limit"],
            item["highalch"]
        )
        labels.append(entry)
    else:
        print(item["id"], "has no limit or highalch key. It will be exlcuded from final output.")

print("Item mappings loaded...")
```

```python
'''Write out item data to a csv'''
outputPath = "data/item_keys.csv"
with open(outputPath, mode='w', newline='') as f:
    # Create csv writer
    csvWriter = csv.writer(f, delimiter = ',')
    
    # Write headers to file
    csvWriter.writerow(headers)
    
    # Write out item data
    for entry in labels:
        csvWriter.writerow(entry)
print(f"Item data written to: {outputPath}")
```
